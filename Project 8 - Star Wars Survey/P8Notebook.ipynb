```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
pd.set_option('display.max_columns',None)
star_wars = pd.read_csv('star_wars.csv',encoding = "ISO-8859-1")
star_wars.info()
```

```python
star_wars = star_wars.loc[pd.notnull(star_wars['RespondentID']),:]
star_wars.head()
```

```python
yes_no = {'Yes':True,'No':False}

for col in star_wars.columns[1:3]:
    star_wars[col] = star_wars[col].map(yes_no)
star_wars.head()
```

```python
import numpy as np
change = {'Star Wars: Episode I  The Phantom Menace':True , np.NaN:False
         ,'Star Wars: Episode II  Attack of the Clones':True
         ,'Star Wars: Episode III  Revenge of the Sith':True
         ,'Star Wars: Episode IV  A New Hope':True
         ,'Star Wars: Episode V The Empire Strikes Back':True
         ,'Star Wars: Episode VI Return of the Jedi':True}
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(change)
star_wars = star_wars.rename(columns={"Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"})
star_wars.head()
```

```python
star_wars = star_wars.rename(columns={"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"})
for e in star_wars.columns[9:15]:
    star_wars[e] = star_wars[e].astype('float')
star_wars.head()
```

```python
name = ['Han Solo','Luke Skywalker','Princess Leia Organa','Anakin Skywalker','Obi Wan Kenobi','Emperor Palpatine','Darth Vader','Lando Calrissian','Boba Fett','C-3P0','R2 D2','Jar Jar Binks','Padme Amidala','Yoda']
star_wars = star_wars.rename(columns={'Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.':name[0],
                            'Unnamed: 16':name[1],
                            'Unnamed: 17':name[2],
                            'Unnamed: 18':name[3],
                            'Unnamed: 19':name[4],
                            'Unnamed: 20':name[5],
                            'Unnamed: 21':name[6],
                            'Unnamed: 22':name[7],
                            'Unnamed: 23':name[8],
                            'Unnamed: 24':name[9],
                            'Unnamed: 25':name[10],
                            'Unnamed: 26':name[11],
                            'Unnamed: 27':name[12],
                            'Unnamed: 28':name[13],
                             })
star_wars.head()
#[15:29]
```

```python
star_wars['Han Solo'].value_counts()
```

```python
change = {'Very favorably':5,
          'Somewhat favorably':4,
          'Neither favorably nor unfavorably (neutral)':3,
          'Somewhat unfavorably':2,
          'Very unfavorably':1,
          'Unfamiliar (N/A)':3}
for e in range(15,29):
    star_wars[star_wars.columns[e]] = star_wars[star_wars.columns[e]].map(change)
star_wars[star_wars.columns[15:29]] = star_wars[star_wars.columns[15:29]].astype('float')
star_wars.head()
```

```python
star_wars.to_csv('Cleaned data.csv')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
star_wars.mean()[9:15].plot.bar(color = (200/255,200/255,200/255))
```

```python
star_wars.sum()[3:9].plot.bar(color = (200/255,200/255,200/255))
```

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
l = ['I','II','III','IV','V','VI']
ax1.bar(range(6),females[females.columns[3:9]].mean(),color=(200/255,200/255,200/255))
ax1.set_title('Female watch')
ax1.set_xticks(range(6))
ax1.set_xticklabels(l)
ax3.bar(range(6),males[males.columns[3:9]].mean(),color=(200/255,200/255,200/255))
ax3.set_title('Female rank')
ax3.set_xticks(range(6))
ax3.set_xticklabels(l)
ax2.bar(range(6),females[females.columns[9:15]].sum(),color=(200/255,200/255,200/255))
ax2.set_title('Male watch')
ax2.set_xticks(range(6))
ax2.set_xticklabels(l)
ax4.bar(range(6),males[males.columns[9:15]].sum(),color=(200/255,200/255,200/255))
ax4.set_title('Male rank')
ax4.set_xticks(range(6))
ax4.set_xticklabels(l)
plt.savefig('Male_vs_Female')
plt.show()
```
